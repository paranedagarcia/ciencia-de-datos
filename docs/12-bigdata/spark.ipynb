# PySpark
PySpark es la API de Python para Apache Spark, que permite procesar grandes volúmenes de datos (Big Data) de forma distribuida y en memoria utilizando la sintaxis sencilla de Python. Combina la potencia de Spark con la facilidad de uso de Pandas, permitiendo realizar ETL, SQL, streaming y Machine Learning en entornos de datos a gran escala.

**Características Principales:**
- Procesamiento Distribuido: Ejecuta tareas en paralelo sobre clusters, ideal para grandes conjuntos de datos.
- DataFrames y SQL: Ofrece una API de alto nivel (DataFrames) similar a Pandas y Spark SQL para consultas estructuradas.
- Interoperabilidad: Utiliza Py4J para comunicarse entre Python y la JVM (Java Virtual Machine) de Spark.
Machine Learning: Incluye librerías para aprendizaje automático a escala. 

**Componentes de la Arquitectura:**
- Driver Program: Coordina el flujo y controla el trabajo.
- Cluster Manager: Gestiona los recursos.
- Worker Nodes/Executors: Ejecutan las tareas y almacenan datos en memoria. 

**Cómo empezar:**
- Requiere Java 8+ y Python instalado.
- Se instala comúnmente mediante pip install pyspark.
- Funciona bien en entornos como Jupyter Notebooks, Databricks o Google Cloud.
- Se inicializa con un SparkSession para crear DataFrames y realizar transformaciones.

**Pasos de instalación:**
1. Instala Java Development Kit (JDK21) (https://www.oracle.com/java/technologies/javase/jdk21-archive-downloads.html)
2. Instala Apache Spark (https://spark.apache.org/downloads.html)
3. Instala Python 3.13 (https://www.python.org/downloads/)


```python
# carga de datos hacia un dataframe de pyspark
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd

spark = SparkSession.builder.appName("Pyspark Tutorial").config("spark.memory.offHeap.enabled","true").config("spark.memory.offHeap.size","10g").getOrCreate()

df = spark.read.option("header", True).option("inferSchema", True).csv("../data/Salaries.csv")
df.show(5, truncate=False)
```
