## 2. Detalles de Geometría y Métricas

La elección de la métrica de distancia es fundamental para el rendimiento del algoritmo. Aquí se detallan las más utilizadas:

### Distancias Fundamentales

#### **Distancia Euclidiana ($L_2$):**
Es la métrica estándar que representa la línea recta entre dos puntos en un espacio n-dimensional. Es, en esencia, el Teorema de Pitágoras aplicado a cualquier número de coordenadas.

$$d(x,y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}$$

El desglose de cada elemento para comprender:
1. Los Puntos ($x, y$)
   - $x$ y $y$: Representan los dos puntos (o vectores) entre los que queremos medir la distancia.
  
   - $d(x,y)$: Es la notación para la "distancia entre el punto $x$ y el punto $y$".

2. Las Dimensiones ($i=1$ hasta $n$)
   - $n$: Es el número total de dimensiones o características (columnas en tu dataset).
   - $i$: Es el índice que recorre cada una de esas dimensiones (primera coordenada, segunda, etc.).
   - $x_i$ y $y_i$: Son los valores específicos de la dimensión $i$ para cada punto. Por ejemplo, si los puntos están en 2D, $x_1$ sería la "X" y $x_2$ sería la "Y".

3. La Operación Interna: $(x_i - y_i)^2$
   - $(x_i - y_i)$: Calcula la diferencia (distancia) entre los dos puntos en una sola dimensión.
   - Elevado al cuadrado $(...)^2$: Se hace por dos razones principales:
     - Eliminar signos negativos: No importa si $x$ es mayor que $y$ o viceversa; la distancia siempre debe sumarse como un valor positivo.
     - Magnitud: Penaliza más las diferencias grandes que las pequeñas.

4. El Sumatorio ($\sum$)
    - $\sum_{i=1}^n$: Este símbolo (Sigma) nos indica que debemos **sumar** todos los resultados obtenidos en el paso anterior para cada una de las $n$ dimensiones.

5. La Raíz Cuadrada ($\sqrt{\dots}$)
    - $\sqrt{\dots}$: Finalmente, aplicamos la raíz cuadrada para "deshacer" el cuadrado del paso 3 y devolver el valor a la unidad de medida original de los datos. Sin la raíz, estaríamos calculando la varianza o la distancia al cuadrado.Resumen del proceso visual:Restas las coordenadas correspondientes.Elevas cada resta al cuadrado.Sumas todos esos cuadrados.Sacas la raíz cuadrada del total.

#### **Distancia de Manhattan ($L_1$):**
Conocida como "distancia del taxista", calcula la suma de las diferencias absolutas. Útil cuando los atributos no son homogéneos. Esta métrica mide la distancia entre dos puntos siguiendo trayectorias perpendiculares (como las calles de una ciudad con manzanas rectangulares). A diferencia de la distancia euclidiana, no permite "atajos" en diagonal.

$$d = \sum_{i=1}^n |x_i - y_i|$$

Desglose de componentes:

$d$: Representa la distancia total calculada entre los dos puntos o vectores.$\sum_{i=1}^n$ 

(Sumatorio): Indica que debemos realizar la misma operación para cada una de las dimensiones (desde la primera $i=1$ hasta la última $n$) y luego sumar todos los resultados.

$| \dots |$ (Barras de Valor Absoluto): Esta es la parte más importante. El valor absoluto asegura que la diferencia sea siempre positiva.
  - Si $x_i = 2$ y $y_i = 5$, la resta es $-3$, pero el valor absoluto lo convierte en $3$.
  - Esto es vital porque las distancias no pueden ser negativas; no tendría sentido que una dimensión "reste" distancia a otra.
$x_i - y_i$: Es la diferencia simple entre las coordenadas del punto $x$ y el punto $y$ en una dimensión específica ($i$).

¿Cómo se interpreta en la práctica?

Imagina que quieres ir del punto A al punto B en una ciudad como Nueva York (Manhattan). No puedes atravesar los edificios en línea recta; tienes que caminar un número determinado de bloques hacia el Norte/Sur y otro número de bloques hacia el Este/Oeste.
1. Restas las posiciones en cada eje.
2. Eliminas el signo negativo (valor absoluto).
3. Sumas todos esos tramos para obtener el recorrido total.

Comparativa rápida con la Euclidiana

| Característica | Distancia Euclidiana (L2) | Distancia de Manhattan (L1) |
| --- | --- | --- |
| Trayectoria | Línea recta (diagonal) | Trayectoria en "escalera" o rejilla |
| Sensibilidad | Muy sensible a outliers (por el cuadrado) | Más robusta frente a valores atípicos |
| Uso común | Geometría física y física clásica | Alta dimensionalidad y GPS urbano |

#### **Similitud de Coseno:**
Mide el ángulo entre dos vectores; muy usada en minería de textos para comparar documentos.

$$\cos(\theta) = \frac{A \cdot B}{\|A\| \|B\|}$$

### Notas sobre Geometrías Específicas

> **DBSCAN y la Geometría:** Este algoritmo no utiliza una métrica para definir un centroide, sino que define puntos "núcleo" si tienen al menos *MinPoints* dentro de un radio . Esto le permite capturar geometrías caprichosas que los métodos basados en centroides (como K-Means) no pueden identificar.

> **Criterio de Ward (Hierarchical):** Este criterio busca minimizar la varianza total dentro de los clusters. En cada paso, se fusionan los clusters que resulten en el menor incremento de la varianza:

$$J_M = \sum_{i=1}^k \sum_{x \in C_i} ||x - m_i||^2$$

> **Actualización en SOM:** El ajuste de pesos en los mapas auto-organizados sigue la fórmula:
$$w_i(t+1) = w_i(t) + f_i(t)[d(t) - w_i(t)]$$



**Implementación de Distancias en Python**

A continuación, calculemos la distancia entre dos puntos en un espacio 2D para observar cómo difieren los resultados según la métrica utilizada.

```python
import numpy as np

# Definimos dos puntos en un espacio 2D
# Punto A (x1, x2) y Punto B (y1, y2)
punto_a = np.array([1, 2])
punto_b = np.array([4, 6])

def calcular_distancias(p1, p2):
    # 1. Distancia Euclidiana (L2)
    # sqrt( sum( (x - y)^2 ) )
    euclidiana = np.sqrt(np.sum(np.square(p1 - p2)))
    
    # 2. Distancia de Manhattan (L1)
    # sum( |x - y| )
    manhattan = np.sum(np.abs(p1 - p2))
    
    return euclidiana, manhattan

dist_e, dist_m = calcular_distancias(punto_a, punto_b)

print(f"Punto A: {punto_a}")
print(f"Punto B: {punto_b}")
print("-" * 30)
print(f"Distancia Euclidiana (L2): {dist_e:.2f}")
print(f"Distancia de Manhattan (L1): {dist_m}")
```

**¿Por qué los resultados son diferentes?**

Al ejecutar el código anterior, notarás lo siguiente:

1. **Distancia Euclidiana** ($\approx 5.0$): Es la longitud del desplazamiento directo (la hipotenusa del triángulo imaginario que forman los puntos). Es el camino más corto posible.

2. **Distancia de Manhattan** ($7.0$): Es la suma de los catetos (3 unidades en el eje X + 4 unidades en el eje Y). Siempre será igual o mayor que la euclidiana.

**¿Cuándo usar cada una en Clustering?**

  - **Usa Euclidiana** si tus datos son densos y las dimensiones están en la misma unidad (ej. coordenadas geográficas reales). Es la que usa K-Means por defecto.

  - **Usa Manhattan** si tienes muchas dimensiones (la "maldición de la dimensionalidad") o si tus datos tienen valores extremos (outliers), ya que no eleva las diferencias al cuadrado y, por tanto, no las magnifica.

**Implementación de K-Means con el Dataset "Wine"**

Para que el clustering funcione correctamente, debemos recordar que la Distancia Euclidiana es sensible a la escala. Por ejemplo, la característica "Magnesio" tiene valores mucho más altos que el "Alcohol". Si no normalizamos, el Magnesio dominará el cálculo de la distancia.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 1. Cargar el dataset
wine = load_wine()
data = wine.data
feature_names = wine.feature_names

# 2. Escalar los datos (Crucial para que la distancia sea equitativa)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 3. Aplicar K-Means
# Sabemos que hay 3 tipos de vino, así que usaremos K=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(data_scaled)

# 4. Reducción de dimensionalidad (PCA) para visualización
# Como no podemos ver 13 dimensiones, reducimos a 2
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data_scaled)

# 5. Visualización de los resultados
plt.figure(figsize=(10, 7))
scatter = plt.scatter(data_2d[:, 0], data_2d[:, 1], c=clusters, cmap='viridis', s=60)
plt.title('Clustering K-Means en Dataset Wine (Proyectado en 2D)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(scatter, label='Cluster asignado')
plt.grid(True, alpha=0.3)
plt.show()

print(f"Inercia del modelo (Suma de distancias cuadradas): {kmeans.inertia_:.2f}")
```

```python
# Ejemplo completo de clustering usando el dataset "wine"

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt

# Cargar dataset y crear DataFrame
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='true_class')
df = pd.concat([X, y], axis=1)

# Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reducción a 2D para visualización
pca = PCA(n_components=2, random_state=0)
X_pca = pca.fit_transform(X_scaled)
df_vis = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_vis['true_class'] = y

# Determinar k con el método del codo (elbow) y calcular silhouette para k>=2
inertias = []
sil_scores = []
ks = range(1, 7)
for k in ks:
    km = KMeans(n_clusters=k, random_state=0, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)
    if k >= 2:
        sil_scores.append(silhouette_score(X_scaled, km.labels_))
    else:
        sil_scores.append(np.nan)

# Elegir k=3 (conocimiento previo del dataset) y ajustar KMeans
k_opt = 3
kmeans = KMeans(n_clusters=k_opt, random_state=0, n_init=10).fit(X_scaled)
labels_km = kmeans.labels_

# Score de silhouette
sil = silhouette_score(X_scaled, labels_km)

# Matriz de confusión entre clases reales y clusters (puede necesitar relabeling)
cm = pd.DataFrame(confusion_matrix(y, labels_km),
                  index=[f"true_{i}" for i in data.target_names],
                  columns=[f"cluster_{i}" for i in range(k_opt)])

# DBSCAN como alternativa (parámetros ejemplo)
db = DBSCAN(eps=0.9, min_samples=5).fit(X_scaled)
labels_db = db.labels_
n_clusters_db = len(set(labels_db)) - (1 if -1 in labels_db else 0)

# Visualizaciones
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(ks, inertias, '-o')
plt.title('Codo (Inertia) para KMeans')
plt.xlabel('k')
plt.ylabel('Inertia')

plt.subplot(2, 2, 2)
plt.plot(ks, sil_scores, '-o')
plt.title('Silhouette score por k')
plt.xlabel('k')
plt.ylabel('Silhouette')

plt.subplot(2, 2, 3)
sns.scatterplot(x='PC1', y='PC2', hue=labels_km, palette='tab10', data=df_vis, legend='full', s=60)
centers_pca = pca.transform(scaler.transform(pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)))
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='black', s=100, marker='X', label='centros')
plt.title(f'KMeans (k={k_opt}) en espacio PCA')

plt.subplot(2, 2, 4)
sns.scatterplot(x='PC1', y='PC2', hue='true_class', palette='viridis', data=df_vis, legend='full', s=60)
plt.title('Clases reales (wine) en espacio PCA')

plt.tight_layout()
plt.show()

# Resultados resumidos
print(f"KMeans elegido k={k_opt} -> silhouette_score = {sil:.3f}")
print("\nMatriz de confusión (true_class vs cluster):")
print(cm)
print(f"\nDBSCAN encontró {n_clusters_db} clusters (label -1 = ruido).")
```

**Conceptos Clave Aplicados**

**Preprocesamiento:** Usamos StandardScaler para que cada característica tenga media 0 y varianza 1. Sin esto, la fórmula $\sqrt{\sum (x_i - y_i)^2}$ daría un peso desproporcionado a las variables con números grandes.

**Inercia:** El valor que imprime el código al final es la suma de las distancias al cuadrado de cada punto a su centroide. El objetivo de K-Means es minimizar este valor.

**PCA (Análisis de Componentes Principales):** Es una técnica que "resume" las 13 dimensiones originales en las 2 dimensiones más informativas para que podamos ver los grupos en un plano.

¿Cómo saber si $K=3$ es el número correcto?

En este caso lo sabemos por la naturaleza del dataset (3 tipos de vino), pero en la vida real usamos el Método del Codo (Elbow Method), graficando la inercia para diferentes valores de $K$ y buscando el punto donde la mejora deja de ser drástica.

¿Cómo interpretar este gráfico de codo?

**Inercia ($WCSS$):** Matemáticamente, la inercia se define como:

$$WCSS = \sum_{j=1}^k \sum_{x_i \in C_j} ||x_i - \mu_j||^2$$

Donde $\mu_j$ es el centroide del cluster $C_j$.

**El balance:** * Si $K$ es igual al número de puntos, la inercia es 0 (cada punto es su propio cluster), pero esto no es útil.
  - Si $K=1$, la inercia es máxima.

**La decisión:** Observarás que entre $K=1$ y $K=3$ la inercia cae drásticamente. A partir de $K=3$, la pendiente se vuelve mucho más suave. Ese "quiebre" en 3 confirma que agrupar los vinos en 3 categorías es la decisión más eficiente desde el punto de vista matemático.

**Limitaciones**

Aunque el método del codo es excelente, a veces el "codo" no es tan claro (curva muy suave). En esos casos, se recomienda complementar con el Análisis de Silueta (Silhouette Score), que mide qué tan bien separado está cada punto de los clusters vecinos.

### Silhouette score

El Análisis de Silueta es una métrica más refinada que el Método del Codo. Mientras que la inercia solo mide qué tan "apretados" están los puntos dentro de un cluster, el coeficiente de silueta mide qué tan cerca está un punto de su propio cluster en comparación con los otros clusters.

El coeficiente oscila entre -1 y 1:

- Cercano a 1: El punto está muy bien clasificado.

- Cercano a 0: El punto está en el límite entre dos clusters.

- Cercano a -1: El punto probablemente pertenece al cluster vecino.

El silhouette score mide qué tan bien está asignado un punto a su clúster comparando la distancia media dentro del clúster propio y la distancia media al clúster más cercano. Para un punto i:

$$s(i) = (b(i) - a(i)) / max(a(i), b(i))$$

- a(i): distancia media de i a otros puntos de su propio clúster (cohesión).
- b(i): mínima distancia media de i a todos los puntos de cualquier otro clúster (separación).

**Interpretación:**
- s ≈ 1: buena asignación (muy cohesivo y bien separado).
- s ≈ 0: punto en la frontera entre clústeres.
- s < 0: posible mala asignación (más cercano a otro clúster).

El silhouette score del conjunto suele ser el promedio de s(i) sobre todos los puntos.

**¿Por qué es mejor este método?**

A diferencia del Método del Codo, donde buscamos un "quiebre" visual, en el gráfico de silueta buscamos el **punto más alto** (el máximo global).

Si el punto más alto está en $K=3$, tenemos una confirmación matemática doble:
- **Inercia:** El modelo es eficiente (poca dispersión interna).

- **Silueta:** Los grupos están bien separados entre sí (poca ambigüedad).


**Visualización de los Perfiles de Silueta (K=3)**

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

# Configuramos el modelo para K=3
k_clusters = 3
kmeans = KMeans(n_clusters=k_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(data_scaled)
silhouette_avg = silhouette_score(data_scaled, cluster_labels)

# Calculamos los valores de silueta para cada muestra
sample_silhouette_values = silhouette_samples(data_scaled, cluster_labels)

fig, ax1 = plt.subplots(1, 1, figsize=(8, 6))

y_lower = 10
for i in range(k_clusters):
    # Agrupamos los valores de silueta de las muestras del cluster i y los ordenamos
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / k_clusters)
    ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    # Etiquetar los clusters en el centro
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

ax1.set_title("Gráfico de Silueta por Cluster (K=3)")
ax1.set_xlabel("Coeficiente de Silueta")
ax1.set_ylabel("ID del Cluster")

# Línea vertical para el promedio
ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
ax1.set_yticks([]) # Limpiar los ejes Y
plt.show()

print(f"Puntaje de silueta promedio: {silhouette_avg:.3f}")
```

Este gráfico muestra el coeficiente de silueta de cada muestra individualmente, agrupadas por cluster.

Es la "prueba de fuego" para el clustering porque nos permite ver si un cluster es muy pequeño, si tiene muestras mal clasificadas (valores negativos) o si un cluster es mucho más denso que otro.

**¿Cómo leer este gráfico?**

1. **Grosor de las "Alas":** Representa cuántos datos hay en cada cluster. En el dataset Wine, verás que los 3 clusters tienen un tamaño similar, lo cual es señal de un buen agrupamiento.

2. **Longitud (Eje X):** Cuanto más a la derecha llegue el "ala", más cohesivo es ese cluster.

3. **Línea Roja Punteada:** Es el promedio total. Un buen clustering debe tener la mayoría de sus alas superando esta línea.

4. **Valores Negativos:** Si ves que alguna parte del gráfico entra en la zona negativa (a la izquierda del 0), significa que esos puntos están más cerca de un cluster vecino que del suyo propio (están "perdidos").

## DBSCAN

Implementación de DBSCAN con el Dataset "Wine"

A diferencia de K-Means, aquí no definimos $K$. Definimos $eps$ (la distancia máxima entre dos puntos para ser vecinos) y min_samples (cuántos puntos se necesitan para formar una "isla").

```python
from sklearn.cluster import DBSCAN

# 1. Aplicar DBSCAN
# eps: radio de vecindad. min_samples: mínimo de puntos para ser considerado un 'core point'
dbscan = DBSCAN(eps=2.3, min_samples=5)
clusters_db = dbscan.fit_predict(data_scaled)

# 2. Visualización usando PCA (reutilizando los datos 2D anteriores)
plt.figure(figsize=(10, 7))
unique_labels = set(clusters_db)
# paleta principal
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

# paleta alternativa (RGBA) por si desea usar otra combinación
alt_colors = [
    (0.121, 0.466, 0.705, 1.0),  # azul
    (1.000, 0.498, 0.054, 1.0),  # naranja
    (0.172, 0.627, 0.172, 1.0),  # verde
    (0.000, 0.000, 0.000, 1.0)   # negro (ruido)
]

for k, col in zip(unique_labels, alt_colors):
    if k == -1:
        # El negro se usa para el ruido (outliers)
        col = [0.172, 0.627, 0.172, 1.0]

    class_member_mask = (clusters_db == k)
    xy = data_2d[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=[col], label=f'Cluster {k}' if k != -1 else 'Ruido', s=60)

plt.title('Clustering DBSCAN en Dataset Wine (Proyectado en 2D)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 3. Resumen de resultados
n_clusters_ = len(set(clusters_db)) - (1 if -1 in clusters_db else 0)
n_noise_ = list(clusters_db).count(-1)

print(f"Número estimado de clusters: {n_clusters_}")
print(f"Número estimado de puntos de ruido: {n_noise_}")
```

**¿Por qué los resultados son tan distintos?**

1. **Sensibilidad a los parámetros:** DBSCAN es extremadamente sensible a eps. Si el radio es muy pequeño, todo es ruido; si es muy grande, todo es un solo cluster.

2. **El problema de la densidad variable:** El dataset Wine tiene densidades similares, pero si un grupo fuera mucho más disperso que otro, DBSCAN tendría dificultades para encontrarlos simultáneamente con un solo eps.

3. **Outliers:** Notarás que DBSCAN probablemente marcó algunos puntos como "Ruido" (negro). K-Means, en cambio, los habría "arrastrado" a la fuerza al cluster más cercano, moviendo el centroide y afectando la precisión.

Comparativa final

| Aspecto | K-Means | DBSCAN |
| --- | --- | --- |
| Sensibilidad a parámetros | Requiere K; elección incorrecta afecta resultados | Muy sensible a eps y min_samples: eps pequeño → todo ruido, eps grande → un solo cluster |
| Densidad variable | Asume clusters globulares; falla con densidades muy distintas | Dificultad para capturar clusters con densidades diferentes usando un solo eps |
| Outliers / Ruido | No detecta ruido: los arrastra y desplazan centroides | Identifica y separa ruido (label -1) |
| Forma del cluster | Esférica / globular (centroidal) | Arbitraria, basada en densidad |
| Complejidad / escalabilidad | Rápido, O(n) en la práctica | Más costoso, típicamente O(n log n), escalable pero más lento |

## Clustering jerárquico

La agrupación jerárquica (Hierarchical Clustering o Agglomerative Clustering) es un tipo de aprendizaje no supervisado que busca construir una jerarquía de agrupaciones (clusters), donde la clasificación de un punto de datos no depende de una solución final predefinida, sino de su posición dentro de una estructura de árbol.

A diferencia de los métodos de "agrupación plana" (flat clustering) como K-Means, que dividen los datos en un conjunto de clusters sin relacionarlos entre sí, la agrupación jerárquica crea una jerarquía de clusters que puede visualizarse a través de un dendrograma.

```python
# Clustering jerárquico completo (autocontenido) con dendrograma — dataset: iris

from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# Cargar datos
iris = load_iris()
X_h = pd.DataFrame(iris.data, columns=iris.feature_names)
y_true = pd.Series(iris.target, name='true_class')

# Escalado
scaler_h = StandardScaler()
X_scaled_h = scaler_h.fit_transform(X_h)

# Matriz de linkage (Ward)
linkage_matrix = linkage(X_scaled_h, method='ward')

# Dendrograma
plt.figure(figsize=(12, 5))
dn = dendrogram(linkage_matrix, labels=y_true.values, 
                leaf_rotation=90, leaf_font_size=8, color_threshold=None)
plt.title('Dendrograma (ward)')
plt.xlabel('Índice de muestra (labels = clase real)')
plt.ylabel('Distancia')
# línea de corte para 3 clusters (opcional, basada en maxclust)
plt.axhline(y=8, color='r', linestyle='--', lw=1)  # ajustar si se desea otra separación visual
plt.tight_layout()
plt.show()

# Extraer clusters cortando en número de clusters deseado
n_clusters = 3
labels_h = fcluster(linkage_matrix, t=n_clusters, criterion='maxclust') - 1  # pasar a 0-based

# Alternativa con AgglomerativeClustering (para obtener mismo n_clusters directamente)
agg = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
labels_agg = agg.fit_predict(X_scaled_h)

# Evaluación rápida
sil_h = silhouette_score(X_scaled_h, labels_h) if len(np.unique(labels_h)) > 1 else np.nan
print(f"Silhouette (fcluster) para {n_clusters} clusters: {sil_h:.3f}")

# Conteo por cluster y matriz de confusión vs clases reales
print("\nConteo por cluster (fcluster):")
print(pd.Series(labels_h).value_counts().sort_index())

print("\nCrosstab (true_class vs cluster):")
print(pd.crosstab(y_true, labels_h, rownames=['true_class'], colnames=['cluster_h']))

# Visualización en 2D usando PCA
pca_h = PCA(n_components=2, random_state=0)
X_pca_h = pca_h.fit_transform(X_scaled_h)
df_vis_h = pd.DataFrame(X_pca_h, columns=['PC1', 'PC2'])
df_vis_h['cluster_h'] = labels_h
df_vis_h['cluster_agg'] = labels_agg
df_vis_h['true_class'] = y_true.values

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(data=df_vis_h, x='PC1', y='PC2', hue='cluster_h', palette='tab10', s=60)
plt.title('Clusters (fcluster) en PCA 2D')
plt.legend(title='cluster_h', loc='best')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df_vis_h, x='PC1', y='PC2', hue='true_class', palette='viridis', s=60)
plt.title('Clases reales (iris) en PCA 2D')
plt.legend(title='true_class', loc='best')

plt.tight_layout()
plt.show()
```

### Cluster jerarquico sobre datos de pacientes
El Heart Disease Dataset de la UCI Machine Learning Repository, contiene información clínica de pacientes y es ideal para este tipo de análisis.

- **Fuente**: [UCI Heart Disease Dataset](https://archive.ics.uci.edu/dataset/45/heart+disease)

- **Características**: Edad, sexo, presión arterial, colesterol, resultados de ECG, etc.

- **Objetivo**: Agrupar pacientes en subgrupos basados en sus características fisiológicas y clínicas, sin usar la etiqueta de diagnóstico (aprendizaje no supervisado).

**linkage**:

Determina el método utilizado para calcular la distancia entre los grupos recién formados durante el proceso de fusión.

**method:**

- **'ward':** Minimiza la suma de las diferencias al cuadrado dentro de todos los conglomerados (minimiza la varianza). Tiende a producir conglomerados de tamaños más regulares y, por lo general, es menos susceptible al ruido.

- **'complete':** Utiliza la distancia máxima entre todas las observaciones de los dos conjuntos (vinculación máxima). Este método también se conoce como vinculación máxima.

- **'average':** Utiliza el promedio de las distancias entre todas las observaciones de los dos conjuntos.

- **'single':** Utiliza la distancia mínima entre todas las observaciones de los dos conjuntos (vinculación mínima). Este método es eficiente, pero puede ser sensible al ruido y a los efectos de "encadenamiento".

```python
# Ejemplo autocontenido: Clustering jerárquico sobre Heart Disease (UCI)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# 1) Cargar datos
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
column_names = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
    'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target'
]
df_raw = pd.read_csv(url, names=column_names, na_values='?')

# 2) Imputación y limpieza
imputer = SimpleImputer(strategy='median')
df = pd.DataFrame(imputer.fit_transform(df_raw), columns=df_raw.columns)
df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna().reset_index(drop=True)

# 3) Seleccionar características (excluir 'target')
features = ['age', 'sex', 'cp', 'trestbps', 'chol', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
X = df[features].copy()

print(f"Datos cargados: {X.shape[0]} muestras, {X.shape[1]} características")

# 4) Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5) Linkage (Ward) y dendrograma
linkage_matrix = linkage(X_scaled, method='ward')
plt.figure(figsize=(12, 5))
dendrogram(linkage_matrix, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=10., show_contracted=True)
plt.title('Dendrograma (Ward) - Heart Disease')
plt.xlabel('Muestras (agrupadas)')
plt.ylabel('Distancia')
plt.tight_layout()
plt.show()

# 6) Cortar para obtener clusters
n_clusters = 3
cluster_labels = fcluster(linkage_matrix, t=n_clusters, criterion='maxclust')
df['cluster'] = cluster_labels

# 7) Evaluación y resumen
sil = silhouette_score(X_scaled, cluster_labels) if len(np.unique(cluster_labels)) > 1 else np.nan
print(f"Silhouette score (k={n_clusters}): {sil:.3f}")
print('\nTamaño por cluster:')
print(df['cluster'].value_counts().sort_index())

cluster_summary = df.groupby('cluster')[features].mean()
print('\nMedias por cluster:')
print(cluster_summary.round(2))

# Heatmap de medias (estandarizamos las medias para visualización)
plt.figure(figsize=(10, 5))
sns.heatmap(cluster_summary.T, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Perfil promedio por cluster (Heart Disease)')
plt.ylabel('Características')
plt.xlabel('Cluster')
plt.tight_layout()
plt.show()

# PCA 2D para visualización de clusters
pca = PCA(n_components=2, random_state=0)
X_pca = pca.fit_transform(X_scaled)
df_vis = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_vis['cluster'] = cluster_labels

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_vis, x='PC1', y='PC2', hue='cluster', palette='tab10', s=60)
plt.title(f'Clusters jerárquicos (k={n_clusters}) en PCA 2D')
plt.legend(title='cluster')
plt.tight_layout()
plt.show()
```

- Cluster 1: Características mixtas, tal vez con problemas de estrés inducido.
- Cluster 2: Pacientes más jóvenes con alta frecuencia cardíaca → posiblemente con dolor torácico no grave.
- Cluster 3: Pacientes mayores, con menor frecuencia cardíaca → posible riesgo cardiovascular alto.
