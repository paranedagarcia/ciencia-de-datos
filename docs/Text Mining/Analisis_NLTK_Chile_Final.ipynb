# Análisis de Texto con NLTK
Este cuaderno realiza el análisis del documento de la Fiscalía utilizando NLTK.

```python
import nltk
import PyPDF2
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import networkx as nx

# Nota: Requiere descarga de datos de NLTK en entorno local
nltk.download('punkt')
nltk.download('stopwords')

def extract_text(pdf_path):
    with open(pdf_path, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        return ' '.join([p.extract_text() for p in r.pages])

text = extract_text('../data/Análisis Criminología Fiscalía Chile.pdf')
tokens = word_tokenize(text.lower(), language='spanish')
stops = set(stopwords.words('spanish'))
filtered = [t for t in tokens if t.isalnum() and t not in stops and len(t) > 3]
print(f'Tokens filtrados: {len(filtered)}')
```

```python
wc = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered))
plt.figure(figsize=(10, 5))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
