# Support Vector Machine (SVM)

Una Máquina de Soporte Vectorial (**Support Vector Machinee**) es un modelo fundamental dentro del **Aprendizaje de Máquina Supervisado** (Supervised Machine Learning). Se utiliza principalmente para problemas de clasificación, aunque también puede aplicarse a problemas de regresión. En términos sencillos, su trabajo es encontrar la mejor manera de separar diferentes grupos de datos.

El objetivo principal de SVM es analizar los datos de entrenamiento y reconocer patrones para construir un modelo que pueda asignar nuevos ejemplos a una categoría u otra. Es popular porque maneja eficientemente grandes conjuntos de datos, puede gestionar datos linealmente separables o no lineales, es robusto ante datos atípicos (outliers) y generaliza bien a datos que nunca ha visto.

Imagina que tienes puntos azules y puntos rojos dispersos en un plano (como canicas en una mesa). La SVM busca la línea (o frontera) que no solo separe los puntos, sino que también esté lo más alejada posible de los puntos más cercanos de ambos colores.

La SVM es como una cerca: no solo separa dos grupos de objetos, sino que se coloca exactamente en el punto medio de la distancia más corta entre ellos, maximizando así la seguridad de la separación.

El objetivo principal de una SVM es maximizar el margen.

- Frontera (Hiperplano): La línea divisoria que separa las clases.

- Margen: La distancia que existe entre la frontera y los puntos de datos más cercanos a ella.

- Vectores de Soporte: Son esos puntos de datos más cercanos. ¡Son los cruciales! Si mueves uno de estos puntos, la frontera óptima cambiará.

Esto es lo que hace que la SVM sea tan efectiva: al maximizar el margen, el modelo logra una mejor capacidad de generalización, lo que significa que es menos probable que se equivoque con datos nuevos que nunca ha visto.



**Orígenes**

Las SVM fueron propuestas por primera vez en los años 90 por los científicos [Vladimir Vapnik](https://en.wikipedia.org/wiki/Vladimir_Vapnik) y [Alexey Chervonenkis](https://en.wikipedia.org/wiki/Alexey_Chervonenkis).

Surgieron como una mejora a otros métodos de clasificación, buscando una solución más robusta y eficiente, especialmente cuando los datos son complejos.

**¿Cómo funciona?**

Separación con margen: La SVM busca la línea (o plano, en 3D) que mejor separa dos grupos, pero no cualquier línea: la que deja el mayor espacio posible ("margen") entre ambos grupos.

Ejemplo: Si tienes manzanas (rojas) y naranjas (naranjas) en una mesa, la SVM dibuja una línea que las separa, pero alejada lo más posible de ambas para evitar confusiones.

Kernel Trick (Truco del núcleo): Si los datos no se pueden separar con una línea recta (como un ovillo de hilos mezclados), la SVM usa funciones matemáticas ("kernels") para transformarlos a un espacio donde sí sea posible separarlos.




## 1. 📐 El Hiperplano y las Matemáticas de la SVM

Comencemos con el corazón matemático de la Máquina de Soporte Vectorial.

### El Hiperplano (La Frontera)

Como mencionamos, el **hiperplano** es la línea (o plano en dimensiones superiores) que separa las clases de datos. Matemáticamente, en un espacio bidimensional o superior, el hiperplano se define mediante la siguiente ecuación:

$$w \cdot x + b = 0$$

* $x$: Es un **vector** que representa un punto de datos de entrada (tus características o *features*).
* $w$: Es el **vector de pesos** (o vector normal) y es perpendicular al hiperplano.
* $b$: Es el **término de sesgo** (o *bias*), que define la posición del hiperplano respecto al origen.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/7/72/SVM_margin.png/330px-SVM_margin.png)

*Fuente: https://en.wikipedia.org/wiki/Support_vector_machine*

### El Margen y los Vectores de Soporte

El objetivo de la SVM es encontrar el hiperplano **óptimo**. Este no es solo cualquier hiperplano, sino el que tiene el **margen** más grande.

El margen se define por dos hiperplanos paralelos, uno para cada clase, que "tocan" a los puntos más cercanos. Estos puntos que definen el margen son los cruciales **Vectores de Soporte**.

Para los vectores de soporte, se cumple:

* Clase 1: $w \cdot x_i + b = +1$
* Clase 2: $w \cdot x_i + b = -1$

La distancia geométrica entre estos dos hiperplanos (el margen total) es $\frac{2}{||w||}$. Por lo tanto, el problema de optimización de la SVM es **minimizar el vector de pesos $||w||$** (que es lo mismo que maximizar el margen).


El vector de pesos $w$ en la ecuación $w \cdot x + b = 0$ en realidad define la orientación (la inclinación o ángulo) del hiperplano. Piensa en $w$ como el vector normal (perpendicular) al hiperplano, lo que fija su dirección.El término de sesgo $b$, por otro lado, es el que define la posición del hiperplano, moviéndolo en el espacio para encontrar el mejor lugar de separación.

## 2. 🏥 Ejemplos Reales de uso de SVM

La Máquina de Soporte Vectorial es extremadamente valiosa en situaciones donde la **precisión** es vital y cuando los datos son complejos (es decir, tienen muchas características).

Aquí tienes tres aplicaciones comunes:

* **Detección de Spam (Clasificación de Texto):** 📧 Se entrena la SVM para clasificar correos electrónicos en dos clases: "spam" o "no spam" (ham). Cada palabra o característica del correo es una dimensión en el espacio, y el hiperplano traza la línea divisoria óptima.

* **Reconocimiento de Imágenes y Rostros:** 🖼️ Al analizar imágenes, cada píxel es una característica. La SVM puede manejar las miles de características para distinguir entre, por ejemplo, diferentes personas o clasificar objetos en una escena.

**Reconocimiento de escritura a mano:** Por ejemplo, reconocer dígitos en cheques escaneados o formularios. Entrada: imagen del dígito transformada en un vector de números. Salida: dígitos del 0 al 9.

**Detección de fraudes en tarjetas de crédito:** Donde se entrena con datos como: monto de la compra, país, tipo de comercio, hora, historial del cliente, etc. Y la salida será: transacción normal o fraudulenta. La SVM ayuda a distinguir patrones típicos de compra de patrones sospechosos.

**Clasificación de clientes (marketing):** Se entrena co información como: ingreso, edad, frecuencia de compra, tipos de productos comprados, etc. Y la salida: grupos de clientes (por ejemplo, “alta probabilidad de comprar cierto producto”). Aunque aquí también se usan técnicas de clustering, las SVM sirven cuando hay ejemplos etiquetados (clientes que sí compraron vs los que no).

* **Diagnóstico Médico (Bioinformática):** 🧬 Se utiliza para clasificar muestras, como la identificación de un tumor como maligno o benigno, basándose en la expresión de múltiples genes o marcadores biológicos (miles de características).

El diagnóstico médico y la bioinformática son campos donde la capacidad de la SVM para manejar miles de genes o marcadores simultáneamente es indispensable, superando a menudo a otros modelos en la claridad de su separación en esos espacios de alta dimensionalidad.

En general, las SVM son muy útiles cuando:

- El número de características no es demasiado grande (o se pueden reducir).

- Tenemos un conjunto de datos moderado en tamaño.

- Se requiere un modelo relativamente interpretable (al menos en el caso lineal) y robusto.

**Ventajas de SVM:**

- Efectivo en espacios de alta dimensión y cuando el número de dimensiones es mayor que el número de muestras. Funciona bien incluso cuando la separación entre grupos no es obvia.
- Utiliza solo un subconjunto de puntos de entrenamiento (vectores de soporte), lo que lo hace eficiente en memoria.
- Flexible gracias al uso de diferentes funciones *kernel* para problemas lineales y no lineales. Gracias al kernel trick, puede manejar relaciones no lineales.
- Generalmente robusto frente al sobreajuste, especialmente en problemas bien separados.

**Desventajas de SVM:**

- El entrenamiento puede ser lento y costoso en memoria para conjuntos de datos muy grandes.

- La elección del *kernel* y sus parámetros puede ser compleja y requiere ajuste cuidadoso. Elegir el kernel y ajustar sus configuraciones requiere experiencia.

- Menor interpretabilidad del modelo en comparación con modelos lineales simples o árboles de decisión. A diferencia de un árbol de decisiones, es menos claro cómo toma las decisiones.

- No proporciona directamente probabilidades de clase (aunque se pueden estimar con métodos adicionales).

**¿Qué son las funciones Kernel en SVM?**  

Las **funciones kernel** permiten a las SVM resolver problemas no lineales proyectando los datos originales a un espacio de mayor dimensión donde sí pueden separarse linealmente. El kernel calcula el producto interno entre las imágenes de los datos en ese espacio, sin necesidad de calcular explícitamente la transformación (truco del kernel).

**Principales tipos de kernel:**

- **Lineal:** $K(x, x') = x^T x'$. No transforma los datos, útil si son linealmente separables.

- **Polinómico:** $K(x, x') = (x^T x' + c)^d$. Permite separar datos con fronteras polinómicas.

- **RBF (Radial Basis Function o Gaussiano):** $K(x, x') = \exp(-\gamma \|x - x'\|^2)$. Muy usado, permite separar datos con fronteras complejas y no lineales.

- **Sigmoide:** $K(x, x') = \tanh(\alpha x^T x' + c)$. Similar a una neurona de red neuronal.

La elección del kernel y sus parámetros es clave para el rendimiento de SVM. El kernel RBF es el más común en la práctica para problemas no lineales.

**¿Qué parámetros se ajustan en el kernel RBF?**  

Los principales parámetros del kernel RBF (Radial Basis Function) en SVM son:

- **C (parámetro de regularización):** Controla el equilibrio entre maximizar el margen y minimizar el error de clasificación. Un valor bajo de C permite un margen más amplio pero más errores; un valor alto busca clasificar correctamente todos los ejemplos, pero puede sobreajustar.

- **gamma ($\gamma$):** Determina la influencia de cada muestra de entrenamiento. Un valor bajo de gamma implica que cada punto tiene un alcance amplio (frontera más suave), mientras que un valor alto hace que la influencia sea muy local (frontera más compleja y ajustada a los datos).

El ajuste de estos parámetros es fundamental para el rendimiento del modelo y suele realizarse mediante validación cruzada y búsqueda en malla (*GridSearchCV*).

### El parámetro **C**

El parámetro **C** (o constante de complejidad) en el algoritmo Support Vector Machine (SVM) es el **parámetro de regularización**.

Su propósito fundamental es gestionar el equilibrio o **trade-off** entre dos objetivos contradictorios durante el entrenamiento del modelo:

1.  **Maximizar el margen de separación** (margen suave o flexible).

2.  **Minimizar los errores de clasificación** cometidos en los datos de entrenamiento.

#### Función Principal

El parámetro $C$ actúa **penalizando el término de error en la función de coste** (o pérdida) del modelo,.

#### Influencia en el Modelo

La elección del valor de $C$ afecta directamente la complejidad y la generalización del modelo:

*   **Valores altos de $C$** (Menos Regularización):
    *   El modelo intenta clasificar correctamente la mayor cantidad posible de puntos de entrenamiento.

    *   Se impone una **mayor penalidad por la clasificación errónea** (margen más "duro").

    *   Esto puede llevar a que el modelo se **sobreajuste (overfit)** a los datos de entrenamiento, lo que significa que el modelo se vuelve muy específico para los datos vistos y podría perder su capacidad de generalizar bien a datos nuevos o no vistos,.

*   **Valores bajos de $C$** (Más Regularización):
    *   El modelo prioriza encontrar un **margen de separación más grande y flexible (margen suave)**.

    *   Al ser menos estricto con los errores de clasificación, permite que algunos puntos queden dentro del margen o clasificados incorrectamente si esto resulta en un hiperplano de decisión más simple (con coeficientes cercanos a cero).

    *   Un valor bajo de $C$ ayuda al modelo a generalizar mejor al hacerlo menos sensible a los datos ruidosos, aunque puede resultar en una precisión ligeramente menor en el conjunto de entrenamiento.

### El parámetro Gamma
El parámetro Gamma ($\gamma$) es un hiperparámetro que se utiliza específicamente con los **kernels no lineales** (como RBF o polinomial) en el algoritmo Support Vector Machine (SVM).

Un valor **alto** para Gamma implica lo siguiente en el contexto del modelo:

1.  **Influencia Localizada:** Un Gamma alto significa que solo los **puntos de entrenamiento cercanos** tendrán una influencia significativa en la determinación del hiperplano de decisión. La frontera de decisión se vuelve muy dependiente de la posición de los ejemplos de entrenamiento individuales.

2.  **Aumento de la Complejidad del Modelo:** Un Gamma alto resulta en un **modelo más complejo**,, lo que permite que el SVM trace fronteras de decisión muy intrincadas o irregulares para clasificar cada punto de entrenamiento con precisión.

3.  **Riesgo de Sobreajuste (Overfitting):** La implicación principal de tener un Gamma alto es que **podría causar sobreajuste** (*overfitting*). El sobreajuste ocurre porque el modelo se ajusta demasiado a las particularidades y al ruido del conjunto de entrenamiento, perdiendo su capacidad para generalizar o clasificar correctamente datos que nunca ha visto antes.

En resumen, un Gamma alto crea una frontera de decisión muy estricta y localizada que se adhiere fuertemente a los puntos de entrenamiento inmediatos, aumentando la complejidad del modelo y, por lo tanto, el riesgo de sobreajuste.

**Ejemplo cotidiano**

Imagina que tienes dos tipos de semillas en una mesa: redondas y alargadas. Una SVM:

1. Mide características como tamaño y peso.

2. Dibuja una línea que las separa lo mejor posible.

3. Si llega una semilla nueva, la clasifica según en qué lado de la línea caiga.

Las SVM son como "superreglas" que aprenden a dividir datos de la manera más precisa posible. Son poderosas en tareas de clasificación, pero su rendimiento depende de la calidad de los datos y la configuración elegida.

**Ejemplo Cotidiano: Clasificar Cítricos por Peso y Radio**

Supongamos que queremos clasificar si una fruta es un limón o una naranja basándonos en su peso y su radio. Esto es un problema de clasificación binaria.

1. **Preparación de Datos**: Se recopilan datos de entrenamiento etiquetados (peso, radio, y la etiqueta: Limón o Naranja).

2. **Elección del Modelo**: Utilizamos el clasificador Support Vector Classifier (SVC) de sklearn.

3. **Ajuste (Training)**: Entrenamos el modelo usando las características (peso y radio). Si asumimos que los cítricos son linealmente separables (es decir, podemos dibujar una línea recta en un gráfico 2D para separarlos), podríamos usar un kernel lineal. Si los datos están mezclados, usaríamos el kernel RBF.

---

```python
# EJEMPLO CONCEPTUAL EN PYTHON USANDO SCIKIT-LEARN

# Importamos las herramientas necesarias
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np 
# Nota: En un caso real, necesitaríamos cargar un dataset (p.ej. con pandas)

# --- 1. Simulación de Datos (Peso y Radio, Clases 0/1) ---
# X_data: [Peso (gr), Radio (cm)]
X_data = np.array([
    [150, 4.5], [160, 4.7], [140, 4.3], # Limones (Clase 0)
    [250, 6.0], [260, 6.2], [245, 5.9]  # Naranjas (Clase 1)
])
# Corrección: la etiqueta debe tener la misma longitud que X_data
y_labels = np.array([0, 0, 0, 1, 1, 1])

# Dividir para entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_data, y_labels, test_size=0.3, random_state=42, stratify=y_labels)

# --- 2. Crear y Entrenar el Modelo (Usando Kernel Lineal) ---
# Inicializamos el clasificador SVM
# Definimos el hiperplano con un kernel lineal, y ajustamos la regularización C
modelo_svm = SVC(kernel='linear', C=1.0) 

# Entrenamos el modelo (el algoritmo aprende la frontera óptima)
modelo_svm.fit(X_train, y_train) 

# --- 3. Hacer Predicciones y Evaluar ---
# Predicción para nuevos datos (p.ej. una fruta de 200gr, 5.0cm)
nuevo_citrico = np.array([[200, 5.0]])
prediccion = modelo_svm.predict(nuevo_citrico)

# Si prediccion es [1], es una naranja. Si es [0], es un limón.

# Evaluación del rendimiento del modelo
predicciones_test = modelo_svm.predict(X_test)
precision = accuracy_score(y_test, predicciones_test)

# Mostrar resultados
print(f"Predicción para [200gr, 5.0cm]: {prediccion}") 
print(f"Precisión del modelo: {precision*100:.2f}%") 

# --- 4. Usando Kernel RBF (Para datos no lineales) ---
# Si el modelo anterior fallara o si los datos fueran más complejos:
modelo_rbf = SVC(kernel='rbf', gamma='scale', C=10.0) 
modelo_rbf.fit(X_train, y_train)
```

## Problemas de clasificación

```python
# 1. Importar Librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions # Para visualizar la frontera

# 2. Preparar los Datos (Usaremos un conjunto de datos simple)
# Cargamos el dataset Iris, solo con dos características (para poder graficarlo)
iris = datasets.load_iris()
X = iris.data[:, [0, 2]] # Tomamos la longitud del sépalo y la longitud del pétalo
y = iris.target

# Simplificamos a 2 clases para una clasificación binaria (Clase 0 vs Clase 1)
X = X[y != 2]
y = y[y != 2]

# 3. Entrenar el Modelo SVM
# Creamos el clasificador SVM:
# kernel='linear' indica que la frontera será una línea recta.
svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(X, y)

# 4. Visualizar la Frontera de Decisión y los Vectores de Soporte
# ---------------------------------------------------------------------
# NOTA: C es el parámetro de Regularización. Controla la penalización del error.
# Si C es pequeño, el margen es más grande (más errores permitidos).
# Si C es grande, el margen es más estrecho (menos errores permitidos).
# ---------------------------------------------------------------------

plt.figure(figsize=(10, 6))
# Trazamos la frontera de decisión:
plot_decision_regions(X, y, clf=svm, legend=2) 

# Marcamos los Vectores de Soporte (los puntos más importantes)
support_vectors = svm.support_vectors_
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=150, 
            facecolors='none', edgecolors='k', linewidths=1.5, label='Vectores de Soporte')

plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Longitud del Pétalo (cm)')
plt.title('SVM Lineal y Vectores de Soporte')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# 5. Imprimir la Ecuación (para relacionar con la teoría)
# Los coeficientes (w) y el intercepto (b)
w = svm.coef_[0]
b = svm.intercept_[0]
print(f"\nResultados del Modelo:")
print(f"Vector de Pesos (w): {w}")
print(f"Sesgo (b): {b}")
print(f"Ecuación del Hiperplano: {w[0]:.2f}*x1 + {w[1]:.2f}*x2 + {b:.2f} = 0")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import numpy as np
from sklearn.svm import SVC as SVC_raw

# Ejemplo completo de clasificación con SVM (scikit-learn)
# Usa las variables/imports ya definidos en celdas previas (iris, datasets, plot_decision_regions, etc.)
# Nuevos imports necesarios:
import matplotlib.pyplot as plt

iris = datasets.load_iris()

# 1. Preparar datos (usamos Iris con 2 características para poder graficar fácilmente)
X = iris.data[:, [0, 2]]   # longitud sépalo, longitud pétalo
y = iris.target

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# 2. Pipeline con escalado y SVC
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('svc', SVC(probability=True, random_state=42))
])

# 3. Búsqueda de hiperparámetros con GridSearchCV
param_grid = {
    'svc__kernel': ['rbf', 'linear'],
    'svc__C': [0.1, 1, 10],
    'svc__gamma': ['scale', 0.01, 0.1]  # gamma solo relevante para rbf
}

grid = GridSearchCV(pipe, param_grid, cv=5, n_jobs=-1, verbose=0)
grid.fit(X_train, y_train)

print("Mejor conjunto de hiperparámetros:", grid.best_params_)
print(f"Mejor score CV: {grid.best_score_:.4f}")

# 4. Evaluación en test set
y_pred = grid.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\nAccuracy en test: {acc:.4f}\n")
print("Reporte de clasificación:\n", classification_report(y_test, y_pred, digits=4))

# Matriz de confusión (visual)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicción')
plt.ylabel('Verdadero')
plt.title('Matriz de Confusión')
plt.show()

# 5. Entrenar modelo final con los mejores hiperparámetros (para visualización)
best_params = grid.best_params_.copy()
# Extraer solo parámetros relevantes para SVC
svc_params = {k.split('__')[1]: v for k, v in best_params.items() if k.startswith('svc__')}
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

final_svc = SVC_raw(**svc_params, probability=True, random_state=42)
final_svc.fit(X_train_scaled, y_train)

# 6. Visualizar fronteras de decisión (usar datos escalados para el clasificador entrenado)
X_plot = np.vstack((X_train_scaled, X_test_scaled))
y_plot = np.hstack((y_train, y_test))

plt.figure(figsize=(10, 6))
# plot_decision_regions requiere las características en la misma escala que el clasificador espera.
plot_decision_regions(X_plot, y_plot, clf=final_svc, legend=2)
plt.title('Frontera de Decisión SVM (mejor modelo)')
plt.xlabel('Feature 1 (escalada)')
plt.ylabel('Feature 2 (escalada)')
plt.show()
```

```python
# EJEMPLO DE SUPPORT VECTOR REGRESSOR (SVR)
# VERSION SIMPLIFICADA

import numpy as np
from sklearn import datasets
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, explained_variance_score
from sklearn.utils import shuffle 
# Nota: En un entorno de producción, se recomienda usar también StandardScaler
# para escalar las características de entrada.

# --- 1. Cargar y Preparar Datos ---

# Cargar el conjunto de datos de viviendas (Boston Housing)
data = datasets.fetch_openml(name='boston', version=1, as_frame=True) # Conjunto de 506 muestras y 13 atributos

# Barajar (shuffle) los datos para evitar sesgos en el análisis 
X, y = shuffle(data.data, data.target, random_state=7)

# Dividir el conjunto de datos en entrenamiento (80%) y prueba (20%)
num_training = int(0.8 * len(X))

# Conjunto de Entrenamiento
X_train, y_train = X[:num_training], y[:num_training]
# Conjunto de Prueba
X_test, y_test = X[num_training:], y[num_training:]

# --- 2. Crear y Entrenar el Modelo SVR ---

# Crear el modelo SVR. Utilizamos un kernel lineal (lineal).
# C (Regularización): Parámetro de penalización por error de entrenamiento. 
# epsilon: Define un umbral de error; si la predicción está dentro de esta distancia, 
# no hay penalización.
sv_regressor = SVR(kernel='linear', C=1.0, epsilon=0.1)

# Entrenar el Support Vector Regressor
sv_regressor.fit(X_train, y_train)

# --- 3. Evaluar el Rendimiento ---

# Realizar predicciones sobre el conjunto de prueba (no visto)
y_test_pred = sv_regressor.predict(X_test)

# Calcular métricas de rendimiento (MSE y EVS)
mse = mean_squared_error(y_test, y_test_pred)
evs = explained_variance_score(y_test, y_test_pred)

# Imprimir el rendimiento 
print("\n#### Rendimiento del SVR ####")
print(f"Error Cuadrático Medio (MSE): {round(mse, 2)}")
print(f"Puntuación de Varianza Explicada (EVS): {round(evs, 2)}")

# --- 4. Predecir un Nuevo Punto de Dato ---

# Ejemplo de datos de entrada para una nueva vivienda (basado en 13 atributos)
test_data = [3.7, 0, 18.4, 1, 0.87, 5.95, 91, 2.5052, 26, 666, 20.2, 351.34, 15.27]

# Predicción del precio
predicted_price = sv_regressor.predict([test_data])
print(f"\nPrecio predicho para la vivienda de prueba: {predicted_price}")
```

El Support Vector Regressor (SVR), al igual que el clasificador SVM, requiere la sintonización de parámetros para optimizar su rendimiento.

• **Kernel**: Especifica el tipo de hiperplano (o función) utilizado para modelar la relación. En este ejemplo, se utiliza el 'linear' kernel, asumiendo una relación lineal, aunque el algoritmo permite kernels no lineales como el RBF o polinomial.

• **C (Parámetro de Regularización)**: Este parámetro representa la penalización por el error de entrenamiento. Si aumenta el valor de C, el modelo se ajustará más para adaptarse a los datos de entrenamiento (buscando un margen de error más estricto), pero podría perder su capacidad de generalización (overfitting).

• **Epsilon (ϵ)**: Este parámetro es crucial para la regresión. Especifica un umbral. Si el valor predicho está dentro de esta distancia (ϵ) del valor real, no se aplica penalización por error de entrenamiento.

**Evaluación del Modelo**

Para evaluar el rendimiento de un modelo de regresión, se utilizan métricas que miden la diferencia entre los valores predichos y los valores reales.

• **Error Cuadrático Medio (Mean Squared Error, MSE)**: Mide la precisión promedio de las predicciones del regressor.

• **Puntuación de Varianza Explicada (Explained Variance Score, EVS)**: Mide qué proporción de la varianza total de la variable objetivo es capturada por el modelo. Un EVS de 1.0 representa una predicción perfecta.

El ejemplo muestra que la predicción de **18.52** para el punto de prueba es consistente con el objetivo real de los datos del conjunto de viviendas

## Optimización de **Hiperparámetros**

La optimización de los hiperparámetros **Kernel**, **C (regularización)** y **Gamma ($\gamma$)** en el algoritmo Support Vector Machine (SVM) utilizando `scikit-learn` se realiza de manera sistemática, generalmente a través de una combinación de **Búsqueda de Cuadrícula (Grid Search)** y **Validación Cruzada (Cross-Validation)**.

En la práctica, el ajuste de estos parámetros es un paso crítico y puede requerir experiencia especializada, especialmente si los datos presentan ruido o se superponen.

### Metodología de Optimización: Grid Search

El método más común y riguroso para la selección de hiperparámetros en `scikit-learn` es utilizar la clase `GridSearchCV`. Esta técnica implementa una **búsqueda exhaustiva** al evaluar todas las combinaciones posibles de los valores que el usuario define para cada parámetro de interés.

Para garantizar que el modelo seleccionado funcione bien con datos nuevos (es decir, que **generalice** adecuadamente y no sufra de sobreajuste), la búsqueda de cuadrícula se ejecuta dentro de un esquema de **Validación Cruzada**.

El proceso general de optimización implica:

1.  **Definición de la Cuadrícula:** Crear un diccionario (`param_grid`) que contenga los nombres de los parámetros y una lista de valores a probar para cada uno.

2.  **Entrenamiento y Evaluación:** `GridSearchCV` entrena el modelo para cada combinación de parámetros y evalúa su rendimiento utilizando la precisión promedio obtenida de la validación cruzada (por ejemplo, 5-fold cross-validation).

3.  **Selección del Mejor Modelo:** Finalmente, se elige la combinación de hiperparámetros que produce la **mejor puntuación media de validación cruzada** (`best_score_`).

### 1. Optimización del Kernel

El parámetro `Kernel` es fundamental ya que selecciona el tipo de hiperplano utilizado para separar los datos.

*   **Valores a Probar:** Se deben incluir los diferentes tipos de kernels, como el kernel lineal (`'linear'`), el kernel de función de base radial (`'rbf'`), o el kernel polinomial (`'poly'`), dependiendo de si se requiere una frontera de decisión lineal o no lineal.

*   **Combinación Condicional:** Dado que los kernels no lineales (como `'rbf'`) requieren el parámetro $\gamma$, y el kernel `'linear'` no lo requiere, la cuadrícula de parámetros a menudo se define como una **lista de diccionarios**, donde cada diccionario prueba un tipo de kernel y sus parámetros específicos (si aplica).

### 2. Optimización del Parámetro C (Regularización)

El parámetro $C$ es la **constante de regularización**, que penaliza el término de error en la función de coste y controla el compromiso entre tener un margen de decisión amplio y la correcta clasificación de los puntos de entrenamiento.

*   **Rango de Búsqueda:** $C$ se utiliza para controlar la complejidad del modelo (donde un $C$ alto puede llevar a un sobreajuste). Por lo tanto, se recomienda buscar su valor en una **escala logarítmica** (por ejemplo, valores como 0.001, 0.01, 0.1, 1, 10, 100) para explorar adecuadamente cómo la magnitud de la restricción afecta el modelo.

### 3. Optimización del Parámetro Gamma ($\gamma$)

El parámetro $\gamma$ es específico de los kernels no lineales, y determina la influencia que tienen los puntos de entrenamiento en la definición del hiperplano.

*   **Rango de Búsqueda:** Un valor de $\gamma$ alto hace que solo los puntos cercanos influyan, lo que **aumenta el riesgo de sobreajuste**. Al igual que $C$, el parámetro $\gamma$ debe ser sintonizado en una **escala logarítmica** como parte del diccionario de parámetros para los kernels no lineales.

En resumen, el SVM es sensible a la selección de parámetros, y la sintonización adecuada, que combina la exploración sistemática (Grid Search) con la evaluación rigurosa (Cross-Validation), es esencial para obtener el mejor rendimiento de generalización.

### Nota sobre Preprocesamiento

Es fundamental tener en cuenta que el rendimiento de SVM es muy sensible a la **escala de los datos**. Por esta razón, antes de ejecutar la búsqueda de parámetros, es una práctica común **normalizar o estandarizar** los datos de entrada para que todas las características varíen en un rango similar, lo cual se realiza utilizando métodos como `MinMaxScaler` o `StandardScaler`.

### Explicación del Proceso

El objeto GridSearchCV automatiza el proceso de tunning de hiperparámetros. Durante el proceso de fit:
1. **Iteración y Validación Cruzada**: GridSearchCV entrena el modelo para cada una de las combinaciones de parámetros definidas en param_grid, utilizando validación cruzada (en este caso, 5-Fold, donde los datos de entrenamiento se dividen en 5 subconjuntos para garantizar la fiabilidad de la evaluación).

2. **Ajuste del Modelo**: En cada iteración de la validación cruzada, el pipeline interno garantiza que el escalado de datos (StandardScaler) se aplique solo a los datos de entrenamiento de ese fold específico, previniendo la fuga de información (information leakage).

3. **Selección**: Una vez que se prueban todas las combinaciones, GridSearchCV selecciona automáticamente la combinación de hiperparámetros (best_params_) que resultó en el mejor rendimiento promedio durante la fase de validación cruzada (best_score_).

4. **Modelo Final**: Finalmente, GridSearchCV ajusta un modelo definitivo utilizando los mejores parámetros encontrados y el conjunto completo de datos de entrenamiento inicial (X_train). Este modelo se almacena en el atributo best_estimator_ y es el que se utiliza para hacer predicciones en el conjunto de prueba (X_test)

```python
# TITANIC DATASET SVM EXAMPLE
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

df = pd.read_csv('../data/titanic.csv')
df['sex_num'] = df['Sex'].map({'male': 0, 'female': 1})
# Usar transform para propagar la media de cada grupo a las filas correspondientes
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda grp: grp.fillna(grp.mean())).round().astype(int)

# Formulacion de modelo
X = df[['Age', 'sex_num']].values
y = df['Survived'].values

# validacion cruzada
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.20, 
                                                    random_state=46)

# Modelamiento SVM
svm_titanic = SVC(kernel='linear', C=1.0, gamma='scale', probability=True, random_state=42)
svm_titanic.fit(X_train, y_train)

# Optimizacion de hiperparametros con GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# Crear un Pipeline para estandarizar los datos antes de aplicar SVM
# Esto asegura que todas las características tengan media 0 y varianza 1, lo cual es vital para los SVM con kernel no lineales.
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(random_state=42, probability=True)) 
])

# Definición de la cuadrícula de búsqueda (Grid)
# Usamos notación de doble guion bajo '__' para referenciar parámetros dentro del Pipeline
param_grid = [
    {
        'svm__kernel': ['rbf'],
        'svm__C': [0.1, 1, 10, 100],  # Probando valores de C en escala logarítmica
        'svm__gamma': [0.001, 0.01, 0.1, 1] # Probando valores de gamma
    },
    {
        'svm__kernel': ['linear'],
        'svm__C': [0.1, 1, 10, 100]
    }
]
# El GridSearchCV probará todas las combinaciones posibles: 4x4=16 para rbf, 
# y 4 para linear (total 20 combinaciones).

# Configurar y ejecutar GridSearchCV
# Usamos cv=5, indicando 5-fold cross-validation.
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# El proceso de fitting entrenará 20 combinaciones x 5 folds = 100 modelos.
grid_search.fit(X_train, y_train)

# Extracción de Resultados

print("Mejores parámetros encontrados (en el set de entrenamiento):")
print(grid_search.best_params_)

print("\nMejor puntuación media de validación cruzada (best_score_):")
print(f"{grid_search.best_score_:.4f}")

# Evaluación final en el set de prueba (test set)
y_pred = grid_search.predict(X_test)
final_accuracy = accuracy_score(y_test, y_pred)

print("\nPrecisión en el conjunto de prueba (Test Set Accuracy):")
print(f"{final_accuracy:.4f}")
```

```python
# prediccion
# Nuevo dato: edad=25, sexo=0
nuevo = np.array([[28, 1]])

mejor = grid_search.best_estimator_
pred = mejor.predict(nuevo)

label_map = {0: "No sobrevivió", 1: "Sobrevivió"}
print(f"Predicción (clase): {pred[0]} -> {label_map[int(pred[0])]}")

# Intentar mostrar probabilidades si el estimador las soporta
try:
    probs = mejor.predict_proba(nuevo)[0]
    print(f"Probabilidades (no, sí): {probs}")
except Exception:
    print("El estimador no soporta predict_proba. " \
    "Reentrene el pipeline con SVC(probability=True) para obtener probabilidades.")
```
