# OLS (Ordinary Least Squares)

La regresión lineal **OLS (Ordinary Least Squares)**, o Mínimos Cuadrados Ordinarios, es la técnica más clásica y fundamental para modelar la relación entre una variable de respuesta continua (dependiente) y una o más variables explicativas (independientes). En el ámbito del aprendizaje automático supervisado, este modelo busca aprender una función lineal que permita predecir resultados futuros a partir de datos históricos.

A continuación se detallan sus componentes y funcionamiento:

### Principios Fundamentales
*   **Objetivo de Optimización:** El método OLS busca los parámetros (coeficientes) que **minimicen la suma de los cuadrados de los residuos**. Un residuo es la diferencia vertical entre el valor real observado y el valor predicho por el modelo. Al elevar estos errores al cuadrado, se asegura que las distancias negativas no se cancelen con las positivas y se penalizan más severamente las desviaciones grandes.

*   **Representación Matemática:** El modelo se expresa mediante una ecuación lineal: **$y = \beta_0 + \beta_1x_1 + \dots + \beta_nx_n$**. Aquí, $\beta_0$ es el intercepto o sesgo (el valor de $y$ cuando las $x$ son cero) y los demás coeficientes ($\beta_n$) representan la pendiente o el peso de cada variable en la predicción.

*   **Interpretación Geométrica:** En un espacio de dos dimensiones, el modelo representa una **línea recta**; en tres dimensiones, un **plano**; y en dimensiones superiores, un **hiperplano**. Geométricamente, el ajuste OLS equivale a realizar una **proyección ortogonal** de los datos observados sobre el subespacio definido por las variables de entrada.

### Métodos de Cálculo
Para encontrar los coeficientes óptimos, existen principalmente dos caminos:
1.  **Solución Analítica (Ecuaciones Normales):** Se utiliza álgebra lineal para resolver la ecuación directamente mediante la fórmula **$\theta = (X^T X)^{-1} X^T y$**. Este método es muy rápido para problemas con pocas variables, pero puede ser costoso computacionalmente si el número de características es muy grande.

2.  **Método Iterativo (Descenso de Gradiente):** Es un algoritmo de optimización que ajusta los coeficientes poco a poco para minimizar el error. Es preferible cuando se trabaja con conjuntos de datos masivos que no caben fácilmente en la memoria del ordenador.

### Validación y Supuestos
*   **R-cuadrado ($R^2$):** Es la métrica estándar para evaluar qué tan bien se ajusta el modelo a los datos; representa la proporción de la varianza de la variable de respuesta que el modelo logra explicar.

*   **Supuestos Clave:** Para que los resultados sean fiables, el modelo asume que existe una **relación lineal** entre las variables, que los datos no tienen un ruido excesivo (valores atípicos) y que no hay **multicolinealidad** (variables independientes altamente correlacionadas entre sí).

***

**Analogía:** Imagine que tiene un grupo de personas dispersas en un campo y quiere colocar una cuerda larga y tensa en el suelo de modo que, en promedio, esté lo más cerca posible de todos. La regresión OLS es como si conectara cada persona a la cuerda con un resorte; la cuerda se moverá y girará hasta que la tensión total de todos los resortes estirados (la suma de los errores al cuadrado) sea la más baja posible.
