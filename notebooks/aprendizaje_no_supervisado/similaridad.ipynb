# **Medidas de Proximidad y Similitud**

 

## **Definición Fundamental**

En machine learning, las **medidas de proximidad** cuantifican qué tan cercanos o lejanos son dos objetos en un espacio de características, mientras que las **medidas de similitud** evalúan qué tan parecidos son.

- **Similitud**: Valores altos indican mayor parecido (rango típico: 0 a 1)
- **Disimilitud/Proximidad**: Valores bajos indican mayor cercanía (mínimo 0, sin máximo fijo)

### **1. Para Datos Numéricos (Vectores Continuos)**

Para conjuntos de datos compuestos por variables numéricas continuas, la selección de una métrica de proximidad es un proceso crítico que depende de la distribución de los datos, la presencia de correlaciones entre características y la sensibilidad deseada a los valores atípicos.

#### **Distancia Euclidiana**
Es la medida de distancia más popular para variables con valores reales continuos, especialmente cuando están en la misma escala. Se calcula como la raíz cuadrada de la suma de las diferencias al cuadrado entre las coordenadas de dos puntos.
La Distancia Euclidiana, siendo la generalización de la distancia Pythagórica, es quizás la más intuitiva y ampliamente utilizada. Su fórmula:
$$
d(p, q) = \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2}
$$
​
 captura la separación espacial absoluta entre dos puntos en un espacio n-dimensional. Su principal fortaleza radica en su simplicidad y su capacidad para modelar la noción humana de "cercanía". Sin embargo, esta misma fuerza es también su debilidad fundamental: la Distancia Euclidiana es extremadamente sensible a la escala de las características 18. Si una característica tiene una varianza mucho mayor que otras (por ejemplo, altura en metros versus peso en kilogramos), dominará la suma de las diferencias al cuadrado, inflando su contribución a la distancia total y distorsionando el resultado. Por esta razón, la estandarización o normalización de los datos es un paso previo indispensable antes de aplicar la Distancia Euclidiana en cualquier escenario real.

**Contexto**: Espacios geométricos "planos", cuando todas las dimensiones tienen igual importancia.
- **Algoritmos**: K-NN, K-Means, DBSCAN, HDBSCAN
- Es fundamental en el algoritmo de K-Means, donde se utiliza para estimar la cohesión o "rigidez" de un cluster
- En la construcción de modelos, si se utiliza esta distancia en los cálculos, a menudo es necesario escalar las características previamente para que ninguna variable con un rango mayor domine los resultados
- **Limitación**: Sensible a escalas diferentes

#### **Distancia Manhattan (L1)**
Una alternativa robusta a la Distancia Euclidiana es la Distancia de Manhatán también conocida como distancia de Bloque Urbano (City Block Distance), se calcula como la suma de los valores absolutos de las diferencias entre las coordenadas de los puntos. Aunque también es sensible a la escala, su penalización lineal en lugar de cuadrática para las diferencias grandes la hace notablemente más robusta a los valores atípicos.

$$
d(x,y) = \sum_{i=1}^{n}|x_i - y_i|
$$
**Contexto**: 
- Datos con rutas en cuadrícula (geográficos), cuando los desplazamientos son ortogonales.
- Es útil cuando las variables de entrada no son del mismo tipo o escala (por ejemplo, edad, género, altura), a diferencia de la distancia euclidiana.
- **Ventaja**: Menos sensible a outliers que la euclidiana

#### **Distancia de Mahalanobis**
Es una generalización de la distancia euclidiana, pero tiene en cuenta la covarianza de las variables para compensar si están correlacionadas, lo que resulta en una forma ponderada de la distancia euclidiana.

Se utiliza en algoritmos como k-NN y en el clasificador de distancia mínima de Mahalanobis, especialmente cuando se asume que los datos en cada clase siguen una distribución gaussiana con una matriz de covarianza común.
$$
d(x,y) = \sqrt{(x-y)^T * S^{-1} * (x-y)}
$$
Donde $S^{-1}$ es la matriz inversa de la covarianza. Esto equivale a transformar los datos para eliminar las correlaciones y luego calcular la Distancia Euclidiana en el nuevo espacio 21. Como resultado, la Distancia de Mahalanobis mide la distancia entre dos puntos en términos de desviaciones estándar dentro de la distribución de datos, lo que la hace invariante a transformaciones lineales y cambios de escala. Es particularmente útil para la detección de anomalías multivariadas, ya que puede identificar puntos que son atípicos en relación con la estructura de correlación del conjunto de datos

**Contexto**: 
- Datos correlacionados, con diferentes varianzas en dimensiones.
- **Considera** la estructura de covarianza de los datos

### **2. Para Datos Binarios y Categóricos**

Las medidas de proximidad se diversifican significativamente cuando se trata con datos que no son numéricos, sino que consisten en atributos binarios (presente/ausente) o categóricos. En estos contextos, el concepto de "distancia" geométrica deja de ser aplicable y surgen métricas diseñadas específicamente para evaluar la superposición o la correspondencia entre conjuntos de características.

#### **Distancia de Hamming**
```python
d(x,y) = número de posiciones donde x_i ≠ y_i
```
**Contexto**: 

- Cadenas binarias, corrección de errores, genómica.

#### **Similaridad de Jaccard**

Se utiliza para conjuntos de datos con atributos binarios, donde se ignora la frecuencia de no ocurrencia. La Similaridad Jaccard es, sin duda, la métrica de referencia para este tipo de datos. Definida como la cardinalidad de la intersección dividida por la cardinalidad de la unión de dos conjuntos, esta medida cuantifica la proporción de elementos compartidos entre dos conjuntos respecto al número total de elementos únicos presentes en cualquiera de ellos 1216. Su rango va de 0 (cuando los conjuntos son disjuntos) a 1 (cuando son idénticos) 1046. Una de sus características más importantes es que ignora explícitamente la coincidencia de ausencias (los pares 0-0), centrándose únicamente en la superposición de atributos positivos (los pares 1-1). 

Esta propiedad la hace particularmente adecuada para datos binarios asimétricos y dispersos, donde la ausencia de un atributo es un evento común y no informativo, mientras que su presencia es significativa.

```python
J(A,B) = |A ∩ B| / |A ∪ B|
```

**Contexto de uso:**
   - Es adecuada para comparar vectores, especialmente en el análisis de texto y para datos binarios.
   - Se usa en el filtrado colaborativo de sistemas de recomendación para medir la similitud entre los "gustos" de los usuarios, sin importar la magnitud de las valoraciones, sino su dirección.
   - datos de presencia/ausencia.
   - análisis de mercado para identificar productos comprados en común por clientes.
   - la detección de plagio comparando conjuntos de palabras en documentos.
   - la evaluación de la precisión de las predicciones de detección de objetos en imágenes

### Hamming distance

La Distancia de Hamming (Hamming Distance, en inglés) es una medida de disimilitud entre dos secuencias de igual longitud, que cuenta el número de posiciones en las que los símbolos correspondientes son diferentes.

Es una métrica fundamental para datos binarios o categóricos discretos, ya que mide el número de posiciones en las que los símbolos correspondientes difieren en dos cadenas de igual longitud. A diferencia de Jaccard, que opera sobre conjuntos, la Hamming Distance requiere que las representaciones sean de la misma longitud, lo que la hace adecuada para comparar secuencias codificadas, como las cadenas de ADN o los códigos de corrección de errores.
Dadas dos cadenas (o vectores) $x$ e $y$ de la misma longitud $n$, la distancia de Hamming se define como:
$$
d_H(x,y) = \sum_{i=1}^{n} (1_{xi \neq yi})
$$
​
Donde $1_{xi \neq yi}$ es una función indicadora que vale:

- **1** si $x_i \neq y_i$
- **0** si $x_i = y_i$
 
En palabras simples: cuenta cuántas posiciones difieren entre las dos secuencias.

**Ejemplo 1:** Cadenas de texto
- x = "karolin"
- y = "kathrin"
Comparación letra por letra:
```
k a r o l i n
k a t h r i n
↑   ↑ ↑     → 3 diferencias
```
Distancia Hamming = 3

**Aplicaciones principales**
1. Teoría de la información y codificación
   - Detectar y corregir errores en transmisiones de datos (códigos de Hamming).
   - Diseñar códigos con distancia mínima de Hamming para tolerancia a fallos.

2. Bioinformática
   - Comparar secuencias de ADN, ARN o proteínas (cuando tienen la misma longitud).
   - Medir mutaciones puntuales.

3. Aprendizaje automático
   - Medir similitud entre vectores binarios o categóricos (ej.: características "sí/no").
   - Usada en algoritmos como k-NN para datos discretos.

4. Criptografía y seguridad
   - Evaluar la diferencia entre claves o hashes.


<br>

En conclusión, para datos binarios y categóricos, la elección de la métrica de proximidad debe reflejar la naturaleza de los atributos. La Similaridad Jaccard es el estándar de oro para conjuntos y datos binarios dispersos donde la superposición es lo que importa. La Distancia Jaccard ofrece una alternativa formal. La Hamming Distance es ideal para secuencias de igual longitud. Y para datos mixtos, una estrategia de codificación y normalización permite adaptar métricas numéricas como la Euclidiana. Utilizar una métrica inapropiada, como aplicar la Similaridad Coseno a datos binarios, sería incorrecto conceptualmente, ya que su función de normalización está diseñada para vectores reales y no tiene sentido en un espacio discreto.

### **3. Para Texto y Vectores Dispersos**
El análisis de datos textuales y otros espacios vectoriales de alta dimensión presenta un desafío único para las medidas de proximidad. En estos dominios, los datos suelen ser intrínsecamente dispersos (sparse) y las características (palabras, tokens) carecen de una escala numérica inherente y comparativa. En tales escenarios, las métricas tradicionales como la Distancia Euclidiana pierden su efectividad y pueden incluso conducir a conclusiones erróneas. La razón fundamental de este fracaso es el "problema de la dimensionalidad", donde en espacios de alta dimensión, la mayoría de los pares de vectores tienden a estar casi ortogonales, y las diferencias en magnitud (longitud del vector) a menudo carecen de significado semántico.

#### **Similaridad del Coseno**

Mide la similaridad como el ángulo entre dos vectores en un espacio multidimensional. Un ángulo menor entre vectores indica una mayor similaridad

$$
cos(θ) = (A·B) / (||A|| * ||B||)
$$

**Contexto**: 
- Procesamiento de lenguaje natural, sistemas de recomendación, cuando la magnitud no importa.
- **Útil para** documentos de diferentes longitudes

#### **Distancia de Levenshtein**
La Levenshtein Distance, o distancia de edición, cuantifica la similitud entre dos cadenas contando el número mínimo de operaciones de inserción, borrado o sustitución de caracteres necesarias para transformar una cadena en otra. Es fundamental en tareas como corrección ortográfica y análisis de secuencias biológicas, pero no se aplica a representaciones de documentos como vectores de características. 

Mide la similitud entre dos secuencias (a menudo texto) calculando el número mínimo de ediciones de un solo carácter (inserciones, eliminaciones o sustituciones) necesarias para transformar una cadena en la otra. Es un método menos robusto para la comparación de textos porque es costoso y sensible al reordenamiento de palabras.

```python
d(x,y) = mínimo número de ediciones (inserción, eliminación, sustitución)
```
**Contexto**: 
- Corrección ortográfica, alineación de secuencias biológicas.

## **Consideraciones Prácticas**

1. **Normalización**: Muchas medidas requieren datos normalizados (ej: euclidiana)
2. **Curse of Dimensionality**: En alta dimensión, todas las distancias tienden a parecerse
3. **Selección de medida**: Depende de:
   - Tipo de datos (continuos, binarios, categóricos)
   - Presencia de outliers
   - Importancia relativa de dimensiones
   - Necesidad de invarianza a escala/rotación

## **Elección por Tipo de Problema**

- **Clustering**: Euclidiana (K-Means), Manhattan (K-Medoids), Coseno (textos)
- **Clasificación**: Euclidiana/Manhattan para K-NN
- **NLP**: Coseno para documentos, Levenshtein para corrección
- **Recomendación**: Coseno (contenido), Jaccard (ítems binarios)
- **Detección de anomalías**: Mahalanobis (considera estructura)

La elección correcta de medida de proximidad/similitud es crucial, ya que define fundamentalmente la noción de "parecido" en el espacio de características, afectando directamente el rendimiento de los algoritmos.
