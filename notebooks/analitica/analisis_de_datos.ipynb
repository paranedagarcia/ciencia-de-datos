```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm

df = pd.read_csv('../data/ventas.csv')

# mostrar todas las columnas del DataFrame
pd.set_option('display.max_columns', None)
df.head()
```

```python
# eliminar filas con Edad nula
antes = df.shape[0]
df.dropna(subset=['Edad'], inplace=True)
df.drop('idoperacion', axis=1, inplace=True)
despues = df.shape[0]
print(f"Filas eliminadas: {antes-despues} | Nuevo shape: {df.shape}")
```

```python
# eliminar registros con las columnas '_PRECIO_VENTA_num'Y 'PRECIO_POR_M2' nulo o cero
antes = df.shape[0]
df = df[(df['_PRECIO_VENTA_num'].notna()) & (df['_PRECIO_VENTA_num'] != 0)]
df = df[(df['PRECIO_POR_M2'].notna()) & (df['PRECIO_POR_M2'] != 0)]
despues = df.shape[0]
print(f"Filas eliminadas por PRECIO_VENTA_num y PRECIO_POR_M2: {antes-despues} | Nuevo shape: {df.shape}")  
```

```python
df.info()
```

```python
# elimina todos los registros con valores cero en columna 'PRECIO_VENTA'
antes = df.shape[0]
df = df[df['PRECIO_VENTA'] != 0]
despues = df.shape[0]
print(f"Filas eliminadas con PRECIO_VENTA == 0: {antes-despues} | Nuevo shape: {df.shape}")  
```

```python
# eliminar columnas que inician con 'DIAS' ya que no son relevantes para el análisis
dias_cols = [col for col in df.columns if col.upper().startswith('DIAS')]
df.drop(columns=dias_cols, inplace=True)
print(f"Columnas eliminadas que comienzan con 'DIAS': {dias_cols}")

# eliminar columnas 'PrecioLista' y 'PagoContado' que no son relevantes para el análisis
df.drop(columns=['PrecioLista', 'PagoContado', 'Descuento'], inplace=True)
print("Columnas eliminadas: 'PrecioLista', 'PagoContado', 'Descuento'")
```

```python
# convertir columnas 'METROS','PrecioLista', 'Descuento', 'PagoContado' a numérica, forzando errores a NaN
for col in ['METROS']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
# convertir columnas que inician con 'FECHA' a tipo date 
fecha_cols = [col for col in df.columns if col.lower().startswith('fecha')]
for col in fecha_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
from scipy.stats import pearsonr, spearmanr

# EDA completo sobre 'df' (ventas / cotizaciones de departamentos)
# Nota: aprovecha imports ya presentes en el notebook (pd, sns, plt, stats, sm)

sns.set(style='whitegrid', rc={'figure.figsize':(10,5)})

def eda_report(df):
    print("1) Dimensiones y tipos")
    print("- shape:", df.shape)
    print("- columnas:", len(df.columns))
    display(df.dtypes)

    print("\n2) Valores faltantes (por columna, %)")
    miss = df.isna().mean().sort_values(ascending=False)
    display(pd.DataFrame({'missing_pct': miss}).head(30))

    print("\n3) Estadísticas descriptivas (numéricas)")
    display(df.describe(percentiles=[.01, .05, .1, .25, .5, .75, .9, .95, .99]).T)

    # skewness, kurtosis, zeros
    num = df.select_dtypes(include=['number']).columns.tolist()
    stats_df = []
    for c in num:
        col = df[c].dropna()
        stats_df.append((c, col.skew(), col.kurtosis(), (col==0).sum(), col.isna().sum()))
    stats_df = pd.DataFrame(stats_df, columns=['col','skew','kurtosis','n_zeros','n_missing']).set_index('col')
    display(stats_df.sort_values('skew', key=abs, ascending=False).head(20))

    print("\n4) Columnas categóricas: top frecuencias")
    cat = df.select_dtypes(include=['object','category']).columns.tolist()
    for c in [c for c in ['FORMA_PAGO','ESTADO','PROYECTO','ComunaProyecto','ORIGEN_VENTA'] if c in df.columns]:
        print(f"\n- {c}:")
        display(df[c].value_counts(dropna=False).head(20))

    print("\n5) Variables de fecha (rango & missing)")
    fecha_cols = [c for c in df.columns if 'FECHA' in c.upper() or 'Fecha' in c]
    for c in fecha_cols:
        if pd.api.types.is_datetime64_any_dtype(df[c]):
            s = df[c]
            print(f"{c}: min={s.min()}, max={s.max()}, missing={s.isna().mean():.2%}")
        else:
            print(f"{c}: no es datetime (tipo={df[c].dtype})")

    # Creaciones útiles
    if 'METROS' in df.columns and 'PRECIO_VENTA' in df.columns:
        df['_PRECIO_VENTA_num'] = pd.to_numeric(df['PRECIO_VENTA'], errors='coerce')
        df['PRECIO_POR_M2'] = df['_PRECIO_VENTA_num'] / df['METROS']
        print("\n6) Precio por m2: resumen")
        display(df['PRECIO_POR_M2'].describe(percentiles=[.01,.05,.1,.25,.5,.75,.9,.95,.99]))

    # Correlaciones
    print("\n7) Matriz de correlación (numéricas principales)")
    num_for_corr = ['METROS','_PRECIO_VENTA_num','Edad','DIAS_COTIZACION','DIAS_RESERVA']
    num_for_corr = [c for c in num_for_corr if c in df.columns]
    if num_for_corr:
        corr = df[num_for_corr].corr(method='pearson')
        display(corr)
        plt.figure(figsize=(6,5))
        sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', center=0)
        plt.title('Correlación (Pearson)')
        plt.show()

    # Gráficos univariados relevantes
    print("\n8) Distribuciones (METROS, PRECIO_VENTA, PRECIO_POR_M2)")
    plot_cols = [c for c in ['METROS','_PRECIO_VENTA_num','PRECIO_POR_M2'] if c in df.columns]
    for c in plot_cols:
        s = df[c].dropna()
        if s.empty: 
            continue
        plt.figure(figsize=(10,3))
        plt.subplot(1,3,1)
        sns.histplot(s, bins=60, kde=True)
        plt.title(f'{c} (raw)')
        plt.subplot(1,3,2)
        sns.boxplot(x=s)
        plt.title('boxplot')
        plt.subplot(1,3,3)
        sns.histplot(np.log1p(s[s>0]), bins=60, kde=True)
        plt.title('log1p')
        plt.tight_layout()
        plt.show()

    # Bivariado: METROS vs PRECIO_VENTA con regresión robusta y outliers
    if 'METROS' in df.columns and '_PRECIO_VENTA_num' in df.columns:
        s_x = df['METROS'].dropna()
        s_y = df['_PRECIO_VENTA_num'].dropna()
        ix = s_x.index.intersection(s_y.index)
        plt.figure(figsize=(8,5))
        sns.regplot(x=df.loc[ix,'METROS'], y=df.loc[ix,'_PRECIO_VENTA_num'], scatter_kws={'s':8,'alpha':0.4}, line_kws={'color':'red'})
        plt.xlabel('METROS')
        plt.ylabel('PRECIO_VENTA')
        plt.title('METROS vs PRECIO_VENTA (regresión)')
        plt.show()

        # estadísticos de correlación
        try:
            print("Pearson r:", pearsonr(df.loc[ix,'METROS'], df.loc[ix,'_PRECIO_VENTA_num']))
            print("Spearman rho:", spearmanr(df.loc[ix,'METROS'], df.loc[ix,'_PRECIO_VENTA_num']))
        except Exception as e:
            print("Error calculando correlaciones:", e)

    # Agrupamientos / pivot tables útiles
    print("\n9) Precio por m2 promedio por TIPOLOGIA y por ComunaProyecto (top 20)")
    if 'PRECIO_POR_M2' in df.columns:
        if 'TIPOLOGIA' in df.columns:
            display(df.groupby('TIPOLOGIA')['PRECIO_POR_M2'].agg(['count','median','mean','std']).sort_values('count', ascending=False).head(20))
        if 'ComunaProyecto' in df.columns:
            display(df.groupby('ComunaProyecto')['PRECIO_POR_M2'].agg(['count','median','mean']).sort_values('count', ascending=False).head(20))

    # Tendencias temporales: conteos y precio medio por mes (FECHACOTIZACION preferida)
    print("\n10) Tendencias temporales (FECHACOTIZACION)")
    if 'FECHACOTIZACION' in df.columns and pd.api.types.is_datetime64_any_dtype(df['FECHACOTIZACION']):
        tmp = df.set_index('FECHACOTIZACION').resample('M').agg({'_PRECIO_VENTA_num':['count','median','mean']})
        tmp.columns = ['_'.join(c) for c in tmp.columns]
        display(tmp.tail(12))
        ax = tmp['_PRECIO_VENTA_num_mean'].plot(title='Precio medio por mes (últimos registros)', legend=False)
        ax.set_ylabel('Precio medio')
        plt.show()

    # Outliers: IQR method para precio y metros
    print("\n11) Detección outliers (IQR) en METROS y PRECIO_VENTA")
    outlier_summary = {}
    for c in ['METROS','_PRECIO_VENTA_num','PRECIO_POR_M2']:
        if c in df.columns:
            s = df[c].dropna()
            q1,q3 = s.quantile([0.25,0.75])
            iqr = q3-q1
            low = q1 - 1.5*iqr
            high = q3 + 1.5*iqr
            n_out = ((s < low) | (s > high)).sum()
            outlier_summary[c] = {'low':low,'high':high,'n_outliers':int(n_out),'pct_outliers': n_out/len(s)}
    display(pd.DataFrame(outlier_summary).T)

    # Recomendaciones rápidas
    print("\n12) Recomendaciones rápidas:")
    print("- Revisar alta proporción de nulos en 'METROS' y varias fechas; imputar o filtrar según análisis.")
    print("- Para análisis de precios usar 'PRECIO_POR_M2' y winsorizar o log-transform si conviene.")
    print("- Agrupar por TIPOLOGIA/ComunaProyecto/PROYECTO para comparar valores y detectar anomalías por subgrupos.")
    print("- Verificar registros con PRECIO_VENTA==0 o METROS==0; pueden ser errores o placeholders.")
    print("- Documentar transformaciones y filtrados antes de modelar.")

# Ejecutar reporte
eda_report(df)
```

## **Análisis univariado**

El análisis univariado es el análisis de una sola variable a la vez para resumir sus características principales y comprender su comportamiento. Es la etapa inicial en el análisis de datos: permite describir, visualizar y detectar problemas (valores faltantes, outliers, distribuciones extrañas) antes de estudiar relaciones entre variables.

### Objetivos principales
- Resumir la variable con estadísticas clave.
- Visualizar su distribución para identificar forma, sesgo y valores atípicos.
- Detectar problemas de calidad de datos (missing, errores, outliers).
- Decidir transformaciones o categorizaciones necesarias para análisis posteriores.

### Qué examinar según el tipo de variable
- Variables numéricas:
    - Medidas de tendencia central: media, mediana, moda.
    - Medidas de dispersión: desviación estándar, varianza, rango, IQR (rango intercuartílico).
    - Forma de la distribución: sesgo (skewness), curtosis.
    - Valores atípicos y posibles transformaciones (log, sqrt).
- Variables categóricas:
    - Frecuencias absolutas y relativas (proporciones).
    - Modo, distribución de categorías y rarezas.
    - Reagrupamiento si hay muchas categorías con baja frecuencia.

### Visualizaciones comunes
- Numéricas: histograma, densidad (KDE), boxplot, violin plot.
- Categóricas: barras, tablas de frecuencia, gráficos de pastel (con precaución).
- Complementos: desviación estándar en gráficos, plot de cuantiles (QQ-plot) para evaluar normalidad.

### Pasos prácticos recomendados
1. Identificar el tipo y el rango de la variable.
2. Calcular estadísticas descriptivas básicas.
3. Visualizar la distribución y buscar outliers/missing.
4. Aplicar transformaciones o binning si es necesario.
5. Registrar hallazgos y decisiones para análisis posteriores.

### Consideraciones y errores comunes
- No confiar sólo en la media si la distribución está sesgada; usar mediana.
- Tratar outliers con cuidado: verificar si son errores o información válida.
- Tener en cuenta el tamaño de la muestra: estadísticas en muestras pequeñas son inestables.
- Documentar transformaciones para reproducibilidad.

El análisis univariado es sencillo pero esencial: establece una base sólida para análisis más avanzados (bivariado y multivariado) y para la limpieza y modelado de datos.

## **Análisis bivariado**

El análisis bivariado estudia la relación entre dos variables para entender si están asociadas, cómo se relacionan (dirección, forma y fuerza) y si esa relación es estadísticamente significativa. Es el paso intermedio entre el análisis univariado y el multivariado, y ayuda a generar hipótesis sobre dependencias o efectos entre variables.

### Objetivos principales
- Detectar asociación o dependencia entre dos variables.
- Cuantificar la dirección (positiva/negativa), forma (lineal/no lineal) y fuerza de la relación.
- Probar si la relación es estadísticamente significativa.
- Visualizar la relación para identificar patrones, outliers o subgrupos.

### Tipos de pares y métodos comunes
- **Numérica — Numérica**
    - Visual: scatter plot, líneas de tendencia.

    - Métricas: correlación de Pearson (lineal), Spearman (monótona), covarianza.

    - Modelos: regresión lineal simple.

- **Categórica — Categórica**
    - Visual: tabla de contingencia, barras apiladas/múltiples.

    - Pruebas: chi-cuadrado de independencia, test exacto de Fisher.

    - Medida de fuerza: Cramér's V.

- **Numérica — Categórica**
    - Visual: boxplot, violín, barras con medias y errores.

    - Pruebas: t-test (dos grupos), ANOVA (más de dos), test de Welch.

    - Medidas: diferencia de medias, tamaño del efecto (Cohen's d).

### Pasos prácticos recomendados
1. Revisar calidad de datos (missing, outliers, escalas).
2. Visualizar la relación con la gráfica adecuada.
3. Calcular estadístico descriptivo (correlación, tablas de frecuencia, medias por grupo).
4. Aplicar prueba estadística de significancia.
5. Interpretar resultados considerando plausibilidad y posibles variables confusoras.

### Errores y consideraciones
- Correlación no implica causalidad.
- Confusores pueden crear asociaciones espurias.
- Outliers pueden distorsionar correlaciones.
- Asegurar tamaño muestral suficiente para pruebas.
- Normalidad y homogeneidad de varianzas condicionan algunas pruebas; usar alternativas no paramétricas si es necesario.

### Ejemplos

1) Numérica — Numérica  
     - Ejemplo: METROS vs PRECIO.  
     - Qué hacer: scatter plot + recta de regresión + calcular correlación de Pearson.  
     - Interpretación típica: correlación positiva alta → a mayor superficie, mayor precio; si r ≈ 0.8 sugiere relación fuerte, pero revisar heterocedasticidad y outliers.

2) Categórica — Categórica  
     - Ejemplo: FORMA_PAGO (Contado/Crédito) vs ESTADO (Reservado/No vendido).  
     - Qué hacer: tabla de contingencia, gráfico de barras apiladas, test chi-cuadrado.  
     - Interpretación típica: chi-cuadrado significativo → la distribución de ESTADO depende de la FORMA_PAGO; usar Cramér's V para medir la fuerza.

3) Numérica — Categórica  
     - Ejemplo: EDAD vs PAGO_CONTADO (Sí/No).  
     - Qué hacer: boxplots por grupo, t-test o Mann–Whitney si no normal.  
     - Interpretación típica: diferencia de medianas significativa → un grupo (p. ej. quienes pagan contado) tiene mayor edad promedio.

4) Mixto/práctico aplicado (opcional)  
     - Ejemplo: LATITUD/LONGITUD vs PRECIO por COMUNA (categórica).  
     - Qué hacer: mapa de calor o scatter geoespacial por precio promedio, ANOVA para comparar medias por comuna.  
     - Interpretación típica: zonas con mayor precio muestran cluster espacial; controlar por METROS y TIPOLOGIA para aislar efecto espacial.

Usar estos análisis para formular hipótesis, no para concluir causalidad. Documentar métodos y resultados para reproducibilidad.

```python
# Scatter + regresión
sns.regplot(x='METROS', y='PRECIO_VENTA', data=df, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title('METROS vs PRECIO')
plt.show()

# Correlaciones
x = df['METROS'].dropna()
y = df['PRECIO_VENTA'].dropna()
ix = x.index.intersection(y.index)
print('Pearson:', stats.pearsonr(x.loc[ix], y.loc[ix]))
print('Spearman:', stats.spearmanr(x.loc[ix], y.loc[ix]))

# Regresión lineal sencilla
X = sm.add_constant(x.loc[ix])
model = sm.OLS(y.loc[ix], X).fit()
print(model.summary())
```

2) Categórica — Categórica (FORMA_PAGO vs ESTADO): tabla de contingencia, chi-cuadrado y Cramér's V

```python

import numpy as np

# Tabla de contingencia y gráfico
ct = pd.crosstab(df['FORMA_PAGO'], df['ESTADO'])
plt.figure(figsize=(12,6))
ct.plot(kind='bar', stacked=False)
plt.title('FORMA_PAGO vs ESTADO')
plt.show()

# Chi-cuadrado
chi2, p, dof, exp = stats.chi2_contingency(ct)
n = ct.values.sum()
phi2 = chi2 / n
r, k = ct.shape
cramers_v = np.sqrt(phi2 / min(k-1, r-1))
print('chi2 p-value:', p)
print("Cramér's V:", cramers_v)
```

```python
# Boxplot por grupo
sns.boxplot(x='PRECIO_VENTA', y='EDAD', data=df)
plt.title('EDAD por PAGO_CONTADO')
plt.show()

# Tests
g1 = df[df['PRECIO_VENTA']=='Sí']['EDAD'].dropna()
g2 = df[df['PRECIO_VENTA']=='No']['EDAD'].dropna()

# t-test de Welch
tstat, pval = stats.ttest_ind(g1, g2, equal_var=False)
print('t-test p-value:', pval)

# Si no normal, Mann-Whitney
mw_stat, mw_p = stats.mannwhitneyu(g1, g2, alternative='two-sided')
print('Mann-Whitney p-value:', mw_p)
```

## **Análisis multivariado**
 
El análisis multivariado estudia simultáneamente más de dos variables para entender relaciones, patrones y dependencias entre ellas. A diferencia del análisis univariado o bivariado, considera la interacción entre múltiples variables y permite responder preguntas como: ¿qué conjunto de variables explican mejor una respuesta?, ¿existen grupos con comportamientos similares?, ¿qué variables aportan más información para un modelo predictivo?

### Objetivos principales
- Identificar relaciones y dependencias entre varias variables a la vez.
- Reducir la dimensionalidad sin perder la mayor parte de la información (p.ej. PCA).
- Construir modelos predictivos multivariados (regresión múltiple, clasificación).
- Detectar estructuras ocultas: clusters, factores latentes, interacciones.
- Seleccionar variables relevantes y evaluar su importancia conjunta.

### Métodos comunes
- Reducción de dimensionalidad: PCA, t-SNE, UMAP.
- Modelos predictivos: regresión lineal múltiple, regresión logística, árboles, random forest.
- Agrupamiento: k-means, DBSCAN, clustering jerárquico.
- Análisis de factores y componentes: Análisis factorial, ICA.
- Técnicas para variables mixtas: análisis discriminante, modelos GLM, modelos de mezcla.

### Pasos prácticos recomendados
1. Preparar datos: tratar nulos, escalado y codificación de categorías.  
2. Explorar correlaciones y multicolinealidad (matriz de correlación, VIF).  
3. Aplicar reducción de dimensionalidad si hay muchas variables.  
4. Probar modelos apropiados según la tarea (predicción, clustering, segmentación).  
5. Validar y evaluar (cross-validation, métricas apropiadas).  
6. Interpretar y comunicar hallazgos: coeficientes, importancia de variables, perfiles de cluster.

1) Reducción de dimensionalidad con PCA (explorar estructuras generales)


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

cols = ['METROS', 'PRECIO', 'EDAD']
df_pca = ventas[cols].dropna()

X = StandardScaler().fit_transform(df_pca)
pca = PCA(n_components=2)
pcs = pca.fit_transform(X)

plt.figure(figsize=(7,5))
plt.scatter(pcs[:,0], pcs[:,1], alpha=0.6, s=20)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%})')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%})')
plt.title('PCA: primeros dos componentes')
plt.grid(alpha=0.2)
plt.show()

print('Varianza explicada total (2 comps):', pca.explained_variance_ratio_.sum())
```

**Interpretación:** el gráfico muestra si los datos tienen direcciones principales de variación; la varianza explicada indica cuánto resumen los primeros componentes.

2) Regresión lineal múltiple (predecir PRECIO usando varias variables)

```python
import statsmodels.formula.api as smf

df_reg = ventas[['PRECIO', 'METROS', 'EDAD']].dropna()
model = smf.ols('PRECIO ~ METROS + EDAD', data=df_reg).fit()
print(model.summary())
```


**Interpretación:** revisar coeficientes, p-values y R^2 para entender qué variables contribuyen a explicar el precio manteniendo las demás constantes.

3) Clustering (segmentación de inmuebles por ubicación y precio)


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

df_cl = ventas[['LONGITUD', 'LATITUD', 'PRECIO']].dropna()
X = StandardScaler().fit_transform(df_cl)
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
df_cl = df_cl.assign(cluster=kmeans.labels_)

plt.figure(figsize=(8,6))
sns.scatterplot(x='LONGITUD', y='LATITUD', hue='cluster', size='PRECIO',
                data=df_cl, palette='tab10', alpha=0.7, sizes=(20,200))
plt.title('Clusters geoespaciales con precio')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Interpretación:** los clusters pueden revelar zonas con comportamientos similares de precio y ubicación; útil para segmentación comercial.

4) Clasificación (logística) — ejemplo simple: predecir pago al contado según edad e ingreso


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

df_clf = ventas[['PAGO_CONTADO', 'EDAD', 'INGRESO']].dropna()
df_clf = df_clf[df_clf['PAGO_CONTADO'].isin(['Sí','No'])].copy()
df_clf['PAGO_BIN'] = (df_clf['PAGO_CONTADO'] == 'Sí').astype(int)

X = df_clf[['EDAD', 'INGRESO']]
y = df_clf['PAGO_BIN']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)

clf = LogisticRegression(max_iter=200).fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print('Coeficientes:', dict(zip(X.columns, clf.coef_[0])))
```

**Interpretación:** métricas de precisión/recall y coeficientes indican cómo edad e ingreso se asocian con la probabilidad de pagar al contado.
